```python
with open("../list.json") as f:
    data = json.load(f)
f.close()
    
with open("../counter.json") as g:
    counter = json.load(g)
g.close()

sorted_counter = sorted(counter.items(), key=operator.itemgetter(1))
countries = sorted_counter[-10:]

country_handle = {}
for c in countries:
    country_handle[c[0]] = data[c[0]]
```

```python
def get_2000_tweets(api, screen_name):
    tweets = []
    x = api.user_timeline(screen_name, count=200)
    old_id = x[-1].id-1
    ids = set()
    while(len(tweets) <= 2000 and len(x)>0):
        for each in x:
            if(each.id not in ids):
                ids.add(each.id)
                tweets.append(each.text)
        x = api.user_timeline(screen_name, count=200, max_id=old_id)
        if(len(x) > 0):
            old_id = x[-1].id-1
    return [tweets, old_id]
```

```python
final_data = {}
old_ids = {}
```

```python
for country, screen_name in data.items():
    k = get_2000_tweets(api, screen_name)
    tweets = k[0]
    print(screen_name, len(tweets))
    old_ids[country] = [screen_name, k[1]]
    final_data[country] = [screen_name, tweets]
```
