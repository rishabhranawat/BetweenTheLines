```python
import json
```

```python
with open("mofo_tweets.json") as f:
    data = json.load(f)
f.close()

with open("../aliases.json") as g:
    aliases = json.load(g)
g.close()

countries = []
for c, d in data.items():
    countries.append(c)

revised_aliases = {}
for key, value in aliases.items():
    if(key in countries): revised_aliases[key] = value

print(revised_aliases)
```

```python
def get_mentions_structure(data):
	mentions = {}
	for country, details in data.items():
		mentions[country] = {}
		for c1, details in data.items():
			if c1 != country: mentions[country][c1] = [0,[]]
	return mentions
```

```python
mentions = get_mentions_structure(data)

for country, details in data.items():
    tweets = details[1]
    for tweet in tweets:
        for c1, a in revised_aliases.items():
            if(c1 != country):
                for a1 in a:
                    if(a1.lower() in tweet.lower()):
                        print(a1, tweet)
                        mentions[country][c1][0] +=1
                        mentions[country][c1][1].append(tweet)
                        break
print(mentions)
```

```python
with open("mentions_tweets_mofo_count.json", "w") as f:
    json.dump(mentions, f)
f.close()

# with open("mentions_tweets_count.json") as g:
#     d = json.load(g)
# g.close()
```

```python
def get_counter_structure(data):
    mentions = {}
    for country, details in data.items():
        mentions[country] = 0
    return mentions
```

```python
counter = get_counter_structure(data)

for country, details in mentions.items():
    for c1, d1 in details.items():
        count = d1[0]
        counter[c1] += count
        
```
